1. What is Dimensionality Reduction? Why is it important in machine
learning?
   - Dimensionality Reduction is the process of reducing the number of input variables or features in a dataset while preserving as much important information as possible. It transforms high-dimensional data into a lower-dimensional form using mathematical techniques such as PCA (Principal Component Analysis), t-SNE, LDA, Autoencoders, etc.
       - Reduces Overfitting Too many features can lead to noise and irrelevant information, causing the model to overfit. Reducing dimensions helps the model generalize better.
       - Improves Model Performance Fewer features mean faster training, faster predictions, and more efficient algorithms.
       - Simplifies Visualization High-dimensional data (e.g., 100+ features) can be projected into 2D or 3D for easy visualization and better understanding.
       - Removes Multicollinearity Some techniques like PCA combine correlated features into independent components, improving model stability.
       - Reduces Storage and Computation Cost Less data means lower memory usage and computation time, especially important for large datasets.
       - Enhances Noise Reduction Dimensionality reduction removes irrelevant and redundant features, keeping only the meaningful information.



2. Name and briefly describe three common dimensionality reduction
techniques..
   - Three Common Dimensionality Reduction Techniques:
       1. Principal Component Analysis (PCA)
           - PCA is a linear dimensionality reduction technique that transforms the original features into a new set of uncorrelated variables called principal components. These components capture the maximum variance in the data, allowing you to reduce dimensions while keeping the most important information.
       2. Linear Discriminant Analysis (LDA)
           - LDA is a supervised dimensionality reduction method that finds new axes that maximize the separation between classes. Unlike PCA, which focuses on variance, LDA focuses on improving classification by projecting data onto directions that best discriminate between categories.
       3. t-Distributed Stochastic Neighbor Embedding (t-SNE)
           - t-SNE is a nonlinear technique mainly used for visualizing high-dimensional data in 2D or 3D. It preserves local structure and clusters similar points together, making it useful for understanding patterns in complex datasets such as images or text embeddings.

3. What is clustering in unsupervised learning? Mention three popular
clustering algorithms.
   - Clustering is an unsupervised learning technique used to group similar data points into clusters based on patterns or similarity in their features. The algorithm does not use labeled data. Instead, it automatically discovers structure in the dataset by placing similar objects together and separating dissimilar ones.
   - Clustering helps in customer segmentation, anomaly detection, image grouping, document classification, etc.
   - Three Popular Clustering Algorithms:
      1. K-Means Clustering
           - A centroid-based algorithm that divides data into k clusters by minimizing the distance between data points and their cluster centroids.
      2. Hierarchical Clustering
           - Builds a tree-like structure (dendrogram) by either merging clusters (agglomerative) or splitting them (divisive). It does not require specifying the number of clusters initially.
      3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
           - A density-based method that groups points close to each other and identifies outliers as noise. Useful for clusters of arbitrary shapes.

      5. Data Cleaning
           - Helps remove outliers or incorrect values that could negatively affect model performance.
      6. Safety and Healthcare Monitoring
           - Used to spot unusual patient vital signs or medical sensor readings.

5. List and briefly describe three types of anomaly detection techniques.
   - Here is an exam-ready answer for “List and briefly describe three types of anomaly detection techniques.”
   - Three Types of Anomaly Detection Techniques:
      1. Statistical Methods
           - These techniques assume that normal data follows a specific statistical distribution (e.g., Gaussian/normal distribution). Data points that fall far from the expected range (e.g., using Z-score, IQR, or confidence intervals) are marked as anomalies.
      2. Machine Learning-Based Methods
           - These use algorithms to learn normal patterns from data. Examples include:
              - Isolation Forest - isolates anomalies by randomly partitioning data.
              - One-Class SVM - learns a boundary around normal data; points outside are anomalies.
              - Autoencoders - reconstruct normal data well; high reconstruction error indicates anomalies.
       3. Distance or Density-Based Methods
           - These methods detect anomalies based on how far or isolated a point is from its neighbors. Examples include:
              - k-Nearest Neighbors (kNN) - points far from neighbors are anomalies.
              - DBSCAN - low-density points are flagged as outliers.
              - Local Outlier Factor (LOF) - compares local density of a point to its neighbors.  

6. What is time series analysis? Mention two key components of time series
data.
   - Time series analysis is a statistical technique used to analyze data points collected or recorded over time at regular intervals (daily, monthly, yearly, etc.). Its goal is to understand patterns, trends, and seasonal behavior in the data and to make future predictions (forecasting).
   - Examples include stock prices, weather data, sales data, sensor readings, and economic indicators.
   - Two Key Components of Time Series Data
      1. Trend
           - The long-term upward or downward movement in the data over time. Example: A company's sales steadily increasing over several years.
      2. Seasonality
           - Regular, repeating patterns within specific time periods. Example: Higher ice-cream sales every summer.
   - Optional Components
       - Cyclic Component - fluctuations over long periods not tied to fixed seasons.
       - Random/Irregular Component - unpredictable variations or noise.

7. Describe the difference between seasonality and cyclic behavior in time
series.
   - Difference Between Seasonality and Cyclic Behavior
       1. Seasonality
            - Seasonality refers to regular and predictable patterns that repeat at fixed intervals within a year. These intervals can be daily, weekly, monthly, or yearly.
                - Occurs at fixed, known periods
                - Short-term repetitive pattern
                - Example: Higher retail sales every December; electricity usage increasing every summer.
       2. Cyclic Behavior
            - Cyclic behavior refers to fluctuations in time series data that follow long-term rise and fall patterns, but do NOT occur at fixed intervals.
                - No fixed or regular duration
                - Influenced by economic or business cycles
                - Example: Economic boom and recession cycles, lasting 3-10 years but not on a fixed schedule.

8. Write Python code to perform K-means clustering on a sample dataset.

```python
#answer in code to ques. no. 8

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate a sample dataset (3 clusters)
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.2)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

9. What is inheritance in OOP? Provide a simple example in Python.
   - Inheritance is an Object-Oriented Programming (OOP) concept that allows a new class (called the child or derived class) to acquire the properties and methods of an existing class (called the parent or base class).
   - It promotes code reusability, extensibility, and helps create hierarchical relationships between classes.

```python
#answer in code to ques. no. 9

# Parent Class
class Animal:
    def sound(self):
        return "Some sound"

# Child Class (inherits from Animal)
class Dog(Animal):
    def sound(self):
        return "Bark"

# Using the classes
a = Animal()
d = Dog()

print(a.sound())   # Output: Some sound
print(d.sound())   # Output: Bark
```

10. How can time series analysis be used for anomaly detection?
    - Time series analysis can be used to detect anomalies by identifying unexpected changes or deviations from normal patterns such as trend, seasonality, or expected values over time.
    - Since time series data is ordered chronologically, anomalies appear as sudden spikes, drops, or unusual patterns that do not fit the historical behavior.
    - How Time Series Helps in Anomaly Detection:
       1. Forecasting-Based Approach
            - Models like ARIMA, Exponential Smoothing, or LSTM predict future values. An anomaly is detected when the actual value deviates significantly from the forecast value.
            - Example: If expected sales = 500 units but actual = 900 units → possible anomaly.
       2. Statistical Threshold Methods
            - Apply statistical rules (mean, variance, standard deviation) to detect unusual points. Common methods:
                - Z-score
                - Moving average with confidence intervals
                - IQR (Interquartile Range)

              If a point lies outside thresholds, it is flagged as an anomaly.
       3. Change Point Detection
            - Detects abrupt changes in the structure of time series (mean/variance shifts). Useful for detecting system failures or sudden behavior changes.
       4. Pattern/Seasonality Decomposition
            - Decompose the series into trend, seasonality, and residual components (using STL or classical decomposition). The residual component is analyzed—large residual errors indicate anomalies
            - Examples Where It Is Used
                - Detecting unusual electricity usage
                - Spotting sudden spikes in network traffic (cybersecurity)
                - Identifying abnormal sensor readings in machines
                - Fraud detection in financial transactions

```python
#answer in code to ques. no. 10

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---- Step 1: Create a sample time series ----
np.random.seed(42)
time = pd.date_range(start="2023-01-01", periods=50, freq="D")
values = np.random.normal(100, 5, 50)

# Introduce anomalies
values[10] = 140   # sudden spike
values[30] = 60    # sudden drop

ts = pd.DataFrame({"Date": time, "Value": values})
ts.set_index("Date", inplace=True)

# ---- Step 2: Compute moving average and standard deviation ----
ts["MA_5"] = ts["Value"].rolling(window=5).mean()
ts["SD_5"] = ts["Value"].rolling(window=5).std()

# ---- Step 3: Set anomaly threshold ----
ts["Upper_Bound"] = ts["MA_5"] + 2 * ts["SD_5"]
ts["Lower_Bound"] = ts["MA_5"] - 2 * ts["SD_5"]

# ---- Step 4: Detect anomalies ----
ts["Anomaly"] = (
    (ts["Value"] > ts["Upper_Bound"]) |
    (ts["Value"] < ts["Lower_Bound"])
)

print(ts[ts["Anomaly"] == True])   # print anomaly points

# ---- Step 5: Plot the results ----
plt.figure(figsize=(10,5))
plt.plot(ts.index, ts["Value"], label="Actual Values")
plt.plot(ts.index, ts["MA_5"], label="Moving Average")
plt.fill_between(ts.index, ts["Upper_Bound"], ts["Lower_Bound"],
                 color="lightgray", alpha=0.5, label="Threshold Band")

# Highlight anomalies
plt.scatter(ts.index[ts["Anomaly"] == True],
            ts["Value"][ts["Anomaly"] == True],
            color="red", label="Anomalies", s=80)

plt.legend()
plt.title("Time Series Anomaly Detection Example")
plt.show()
```
